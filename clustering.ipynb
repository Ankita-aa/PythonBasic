```python
# Q1: What is unsupervised learning in the context of machine learning?
# A1: Unsupervised learning involves training a model on data without labeled outputs.
#     The goal is to discover hidden patterns or groupings in the data (e.g., clustering, dimensionality reduction).

# Q2: How does K-Means clustering algorithm work?
# A2: K-Means partitions data into K clusters by:
#     1. Randomly initializing K centroids.
#     2. Assigning each point to the nearest centroid.
#     3. Recalculating centroids based on assigned points.
#     4. Repeating steps 2 and 3 until convergence.

# Q3: Explain the concept of a dendrogram in hierarchical clustering.
# A3: A dendrogram is a tree-like diagram that records the sequences of merges or splits in hierarchical clustering.
#     It helps visualize how clusters are formed at different distance thresholds.

# Q4: What is the main difference between K-Means and Hierarchical Clustering?
# A4: K-Means is a partitional method requiring the number of clusters upfront.
#     Hierarchical clustering builds a hierarchy of clusters without needing the number of clusters beforehand.

# Q5: What are the advantages of DBSCAN over K-Means?
# A5: - DBSCAN can find clusters of arbitrary shapes.
#     - It doesn't require the number of clusters in advance.
#     - It can identify noise points (outliers).

# Q6: When would you use Silhouette Score in clustering?
# A6: Silhouette Score measures how well each point fits within its cluster compared to others.
#     It's used to evaluate clustering quality and select the optimal number of clusters.

# Q7: What are the limitations of Hierarchical Clustering?
# A7: - Computationally expensive for large datasets (O(n^3)).
#     - Once a merge is done, it cannot be undone.
#     - Sensitive to noise and outliers.

# Q8: Why is feature scaling important in clustering algorithms like K-Means?
# A8: K-Means uses Euclidean distance, which is sensitive to feature magnitude.
#     Unscaled features can dominate the distance calculation, skewing results.

# Q9: How does DBSCAN identify noise points?
# A9: DBSCAN marks points as noise if they:
#     - Have fewer than `min_samples` points within the `eps` radius.
#     - Are not reachable from any dense region.

# Q10: Define inertia in the context of K-Means.
# A10: Inertia is the sum of squared distances between each point and its assigned centroid.
#      Lower inertia means tighter clusters, but it doesn't always imply better clustering.

# Q11: What is the elbow method in K-Means clustering?
# A11: The elbow method involves plotting inertia vs. number of clusters.
#      The "elbow" point (where the curve bends) indicates the optimal number of clusters.

# Q12: Describe the concept of "density" in DBSCAN.
# A12: Density refers to the number of points within a specified radius (`eps`).
#      High-density regions are considered clusters.

# Q13: Can hierarchical clustering be used on categorical data?
# A13: Yes, but standard methods like Euclidean distance donâ€™t work well.
#      Alternatives like Hamming distance or Gower distance can be used.

# Q14: What does a negative Silhouette Score indicate?
# A14: It means the sample is assigned to the wrong cluster and is closer to another cluster than its own.

# Q15: Explain the term "linkage criteria" in hierarchical clustering.
# A15: Linkage criteria determine how distances between clusters are calculated.
#      Common types: single, complete, average, ward linkage.

# Q16: Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
# A16: K-Means assumes clusters are spherical and of equal size/density.
#      It may split larger clusters or merge smaller ones incorrectly.

# Q17: What are the core parameters in DBSCAN, and how do they influence clustering?
# A17: - `eps`: Radius of neighborhood.
#      - `min_samples`: Minimum number of points to form a dense region.
#      Larger `eps` merges clusters; smaller `eps` detects finer structures.

# Q18: How does K-Means++ improve upon standard K-Means initialization?
# A18: K-Means++ selects initial centroids more strategically:
#      Each new centroid is chosen with a probability proportional to its squared distance from existing ones.
#      This improves convergence and clustering quality.

# Q19: What is agglomerative clustering?
# A19: Agglomerative clustering is a bottom-up hierarchical clustering approach.
#      It starts with each point as a cluster and merges the closest pairs until one cluster remains.

# Q20: What makes Silhouette Score a better metric than just inertia for model evaluation?
# A20: Inertia only measures compactness, not separation.
#      Silhouette Score considers both how close points are within a cluster and how far they are from others,
#      offering a more balanced evaluation.
```

```python
#24
#Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster
from sklearn.datasets import load_wine

wine = load_wine()

x = pd.DataFrame(wine.data, columns = wine.feature_names)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)

from sklearn.cluster import KMeans

kmean = KMeans(n_clusters=3)
kmean.fit(x)
y_label = kmean.predict(x)

# Print size of each cluster
unique, counts = np.unique(y_label, return_counts=True)
for label, count in zip(unique, counts):
    print(f"Cluster {label}: {count} samples")
```

```python
# 25 Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result

from sklearn.datasets import make_circles
x, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=0)

dd = pd.DataFrame(x , columns= ['x1', 'x2'])

from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = 0.3, min_samples = 5)
dbscan.fit(x)
dd['cluster'] = dbscan.labels_
# Plot the result
plt.figure(figsize=(6, 6))
plt.scatter(dd['x1'], dd['x2'], c=dd['cluster'], cmap='viridis', s=30)
plt.title("DBSCAN Clustering on make_circles Data")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import pandas as pd

# Load the dataset
cancer = load_breast_cancer()
x = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Scale the data using MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)

# Apply KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(x_scaled)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Convert centroids to DataFrame using correct number of columns
centroid_df = pd.DataFrame(centroids, columns=x.columns)

print("Cluster Centroids:")
print(centroid_df)
```
